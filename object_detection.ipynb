```python
from ultralytics import YOLO #Imports the YOLO class from the ultralytics package.
#Functionality: This lets you load YOLOv8 models and run tasks like detection, segmentation, and classification.
import cv2 #
from PIL import Image #Purpose: Imports the Python Imaging Library (PIL).
#Functionality: Allows you to manipulate and display images. Not used directly in this snippet.
import matplotlib.pyplot as plt

# Load pretrained YOLOv8 model (v8n is the smallest and fastest model)
model = YOLO("yolov8n.pt")  # you can also use yolov8s.pt, yolov8m.pt, etc.

# Load an image (upload your image to Colab or use a URL)
image_path = '/content/restaurent_eating.jpg'  # replace with your image
results = model(image_path)   # run detection

# Show the detected objects
results[0].show()  # displays image with boxes
#When you run YOLO on an image, it returns a list called results.
#If you give it one image, it gives you one result, which is at index 0.
#So results[0] is the detection result for your image.
#results[0] accesses the result of the first image.
# Save output image (optional)
results[0].save(filename="output.jpg")
```
